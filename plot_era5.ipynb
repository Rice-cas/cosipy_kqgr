```python
import xarray as xr
import os
import numpy as np
import matplotlib.pyplot as plt

# 定义数据文件夹路径
data_folder = 'data/output'

# 列出所有以 'ERA5' 开头且以 '.nc' 结尾的文件，并按文件名排序
files = sorted([f for f in os.listdir(data_folder) if f.startswith('ERA5') and f.endswith('.nc')])

# 初始化一个列表来存储均值
mean_values = []

# 遍历每个文件并执行操作
for file in files:
    print(file)
    file_path = os.path.join(data_folder, file)
    ds = xr.open_dataset(file_path)
    mean_value = ds.mean(dim=['lat', 'lon']).sum(dim='time')
    mean_values.append(mean_value['MB'])

# 将均值列表转换为 numpy 数组
mean_values = np.array(mean_values)

# 可选：可以查看均值数组
print(mean_values)
```

```python
import numpy as np
import matplotlib.pyplot as plt


# 生成年份（2000-2015年）
years = np.arange(2000, 2024)



# 绘制图形
plt.figure(figsize=(10, 6))
plt.plot(years, mean_values, marker='o', linestyle='-', color='b', label='Temperature Anomaly')

# 添加标题和标签
plt.title('Temperature Anomaly from 2000 to 2015', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Temperature Anomaly (°C)', fontsize=14)

# 设置网格线
plt.grid(True, linestyle='--', alpha=0.6)

# 美化坐标轴
plt.xticks(years, rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.ylim(-20,-7.5)
# 添加图例
plt.legend(fontsize=12)

# 调整布局
plt.tight_layout()

# 显示图形
plt.show()
```
